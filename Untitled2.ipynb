```python
import os
import shutil
import glob

# --- Configuration (UPDATED) ---
# **VERIFY THIS PATH:** It must point to the folder containing your 5 class folders (Pneumonia, TB, etc. AND Normal).
root_dir = 'Data of Disease' 
# Added 'Normal' class
original_classes = ['Pneumonia', 'TB', 'Lung_Cancer', 'Bone_Fracture', 'Normal'] 
standard_classes = ['pneumonia', 'tb', 'lung_cancer', 'bone_fracture', 'normal']

# Output: Standard structure
train_dir = 'dataset/train'
test_dir = 'dataset/test'
# --- End Configuration ---

# Step 1: Root Check
if not os.path.exists(root_dir):
    print(f"❌ ERROR: Root directory '{root_dir}' not found! Check path/spelling.")
    print("Current directory:", os.getcwd())
    print(f"Expected to find subfolders {original_classes} inside this root.")
    # Exit if the root folder is missing
    exit()

print(f"✅ Root directory '{root_dir}' found.")
print("\n=== Combining Dataset to Standard Structure (5 Classes) ===")

# Create standard directories
for dir_path in [train_dir, test_dir]:
    os.makedirs(dir_path, exist_ok=True)
    for class_name in standard_classes:
        os.makedirs(os.path.join(dir_path, class_name), exist_ok=True)

# Function to copy images
def copy_images(split_name, original_classes, standard_classes, root_dir, output_base_dir):
    total_copied = 0
    print(f"\nCopying {split_name.upper()} images...")
    
    for i, orig_class in enumerate(original_classes):
        std_class = standard_classes[i]
        # Assumes structure: root_dir/Original_Class/train or test/images.jpg
        source_split_dir = os.path.join(root_dir, orig_class, split_name)
        dest_dir = os.path.join(output_base_dir, split_name, std_class)
        
        if os.path.exists(source_split_dir):
            copied_count = 0
            # Search for common image file extensions
            image_files = glob.glob(os.path.join(source_split_dir, '*.[jJpPtT][pPnNfF][gG]', '*.[jJpPtT][pPnNfF][gG]')) # .jpg, .jpeg, .png, .tiff etc
            image_files = [f for f in os.listdir(source_split_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff'))]
            
            for img_file in image_files:
                name, ext = os.path.splitext(img_file)
                # Create a unique name
                new_name = f"{name}_{std_class}_{split_name}{ext}" 
                shutil.copy(
                    os.path.join(source_split_dir, img_file),
                    os.path.join(dest_dir, new_name)
                )
                copied_count += 1
            print(f"  - {orig_class} -> {std_class}: {copied_count} {split_name} images copied")
            total_copied += copied_count
        else:
            print(f"  - ❌ ERROR: Source directory for {orig_class} ({source_split_dir}) not found! Please check your folder structure.")
            
    return total_copied

# Step 2 & 3: Copy Train and Test
total_train = copy_images('train', original_classes, standard_classes, root_dir, 'dataset')
total_test = copy_images('test', original_classes, standard_classes, root_dir, 'dataset')

# Step 4: Final Summary
print(f"\n=== Summary ===")
print(f"Total TRAIN images copied: {total_train}")
print(f"Total TEST images copied: {total_test}")
print(f"Grand Total: {total_train + total_test}")
print(f"Standard structure created at: {train_dir} and {test_dir}")

# Verify: Count in new folders
print("\nVerification (New Structure Counts):")
for split in ['train', 'test']:
    split_dir = os.path.join('dataset', split)
    print(f"\n{split.upper()}:")
    for std_class in standard_classes:
        class_dir = os.path.join(split_dir, std_class)
        count = 0
        if os.path.exists(class_dir):
            count = len([f for f in os.listdir(class_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff'))])
        print(f"  - {std_class}: {count} images")

if total_train + total_test > 0:
    print("\n✅ Combine successful! Now run the next cells for training.")
else:
    print("\n⚠️ WARNING: No images were copied. Please check the 'root_dir' path and folder structure.")
print("Original data safe – only copied.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
# Added BatchNormalization for better performance
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization 
from tensorflow.keras.optimizers import Adam
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import itertools # Needed for better class weight handling

# --- Configuration ---
train_dir = 'dataset/train'
test_dir = 'dataset/test'
img_height, img_width = 224, 224
batch_size = 32
epochs = 20 
learning_rate = 0.001
# --- End Configuration ---

# Step 1: Paths Check
if not os.path.exists(train_dir) or not os.path.exists(test_dir):
    raise FileNotFoundError("❌ Dataset directory 'dataset/train' or 'dataset/test' not found! Run the Dataset Combination cell first.")

print("=== Model Training Started ===")

# Step 2: Data Generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1,
    validation_split=0.2 
)
test_datagen = ImageDataGenerator(rescale=1./255)

# Load training data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    color_mode='grayscale',
    shuffle=True
)

# Load validation data
val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    color_mode='grayscale',
    shuffle=True
)

# Load test data
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False, # ESSENTIAL for evaluation metrics
    color_mode='grayscale'
)

# Dynamic Classes
classes = list(train_generator.class_indices.keys())
actual_num_classes = len(classes)
print(f"Detected Classes from Train: {classes} (Total: {actual_num_classes})")
print(f"Train samples: {train_generator.samples}, Validation samples: {val_generator.samples}, Test samples: {test_generator.samples}")

y_true = test_generator.classes 

# Step 3: Class Weights (Handles Imbalanced Data for 5 classes)
if train_generator.samples > 0:
    y_train = train_generator.classes
    # Ensure all classes are represented in unique array
    unique_classes = np.arange(actual_num_classes)
    class_weights = compute_class_weight('balanced', classes=unique_classes, y=y_train)
    class_weight_dict = dict(enumerate(class_weights))
    print("Class weights (used to balance training):", class_weight_dict)
else:
    print("⚠️ WARNING: No training samples found. Cannot compute class weights or train.")
    class_weight_dict = None


# Step 4: Model Definition (CNN Architecture with 5-class output)
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 1)),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    
    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    
    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    
    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(actual_num_classes, activation='softmax') # Dynamic output layer (now 5)
])

model.compile(
    optimizer=Adam(learning_rate=learning_rate),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nModel Summary:")
model.summary()

# Step 5: Train
if train_generator.samples > 0 and actual_num_classes > 1:
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // batch_size,
        epochs=epochs,
        validation_data=val_generator,
        validation_steps=val_generator.samples // batch_size,
        class_weight=class_weight_dict,
        verbose=1
    )

    # Step 6: Evaluate
    print("\n--- Model Evaluation ---")
    if test_generator.samples > 0:
        test_loss, test_acc = model.evaluate(test_generator)
        print(f"\nTest Accuracy: {test_acc * 100:.2f}%\n")

        # Predictions
        test_generator.reset() 
        y_pred = model.predict(test_generator)
        y_pred_classes = np.argmax(y_pred, axis=1)

        # Step 7: Classification Report and Confusion Matrix
        labels = list(range(actual_num_classes))
        print("Classification Report:")
        print(classification_report(y_true, y_pred_classes, labels=labels, target_names=classes, zero_division=0))

        cm = confusion_matrix(y_true, y_pred_classes, labels=labels)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
        plt.title('Confusion Matrix (5 Classes)')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()
    else:
        print("⚠️ WARNING: No test samples found. Skipping final evaluation metrics.")

    # Step 8: Plots
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title('Model Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Model Loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Step 9: Save
    model_save_path = 'multi_disease_xray_model_5class.h5'
    model.save(model_save_path)
    print(f"\n✅ Model saved successfully to: {model_save_path}")
else:
    print("Cannot proceed with training or evaluation as there are not enough classes or training data.")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import os


model_path = 'multi_disease_xray_model_5class.h5'
# UPDATE THIS PATH to an actual image file you want to test
img_path_to_test = r'C:\Users\ASUS\OneDrive\Desktop\jupyter\Heart Diesese prediction\Data of Disease\Pneumonia\train\0a03fbf6-3c9a-4e2e-89ce-c7629ae43a27.png' 
img_height, img_width = 224, 224


classes = ['bone_fracture', 'lung_cancer', 'normal', 'pneumonia', 'tb'] 
# --- End Configuration ---

if not os.path.exists(model_path):
    print(f"❌ ERROR: Model file '{model_path}' not found! Run the training cell first.")
elif not os.path.exists(img_path_to_test):
    print(f"❌ ERROR: Test image file '{img_path_to_test}' not found! Update 'img_path_to_test'.")
else:
    try:
       
        model = load_model(model_path, compile=False) 

        # Load and preprocess the image
        img = image.load_img(img_path_to_test, target_size=(img_height, img_width), color_mode='grayscale')
        img_array = image.img_to_array(img) / 255.0 # Rescale
        img_array = np.expand_dims(img_array, axis=-1)  # Add channel dimension (1 for grayscale)
        img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension (1 image)

        # Make prediction
        predictions = model.predict(img_array)
        
        # Interpret results
        predicted_index = np.argmax(predictions[0])
        predicted_class = classes[predicted_index]
        confidence = predictions[0][predicted_index] * 100

        print(f"\n--- Prediction Result ---")
        print(f"Image Tested: {img_path_to_test}")
        print(f"Predicted Disease: **{predicted_class.upper()}**")
        print(f"Confidence: {confidence:.2f}%")
    
    except Exception as e:
        print(f"An error occurred during prediction: {e}")
```
